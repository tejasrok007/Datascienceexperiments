# 2. Data Cleaning - removing noisy values

```python

mental_health["post_created_day"]=mental_health["post_created"].str[:3]

mental_health["post_created_date_time"]=mental_health["post_created"].str[4:]



mental_health['Dates'] = pd.to_datetime(mental_health['post_created_date_time']).dt.date
mental_health['Time'] = pd.to_datetime(mental_health['post_created_date_time']).dt.time
mental_health.head()
mental_health=mental_health.drop("post_created_date_time",axis=1)

mental_health.head()
```

# 3. Data Transformation - converting numerical attributes to categorical and vice versa/ one hot encoding

```python
# Turn the categories into numbers
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
categorical_features = ["post_created","post_text","post_created_day","Dates","Time"]
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot",
                                   one_hot,
                                   categorical_features)],
                                   remainder="passthrough")

transformed_X = transformer.fit_transform(mental_health)
transformed_X
```

```python
dense_transformed_X = transformed_X.toarray()  # Convert to dense matrix
df_transformed_X = pd.DataFrame(dense_transformed_X)  # Convert to DataFrame
df_transformed_X.head()
```

# 4. Data Transformation - data normalization

```python
#We can use the following code to apply a min-max normalization to each column in the DataFrame:
#1. Min-Max Normalization
df_for_normalization=pd.read_csv("Mental-Health-Twitter.csv")
#Objective: Converts each data value to a value between 0 and 1.
#Formula: New value = (value – min) / (max – min)
# Select numeric columns only
numeric_columns = mental_health.select_dtypes(include=['float64', 'int64']).columns

# Min-max normalization
df_for_normalization[numeric_columns] = (df_for_normalization[numeric_columns] - df_for_normalization[numeric_columns].min()) / (df_for_normalization[numeric_columns].max() - df_for_normalization[numeric_columns].min())
```

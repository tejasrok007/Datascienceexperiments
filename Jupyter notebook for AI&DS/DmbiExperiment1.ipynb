```python
df.dropna(inplace=True)
```

```python
NAN = [(c, df[c].isna().mean()*100) for c in df]
NAN = pd.DataFrame(NAN, columns=["column_name", "percentage"])
NAN.sort_values("percentage", ascending=False)
```

```python
df.isnull().sum()
```

```python
df.drop(df.columns[df.columns.str.contains('Unnamed',case=False)],axis = 1,inplace = True)
df.drop(df.columns[df.columns.str.contains('post_id',case=False)],axis = 1,inplace = True)
df.drop(df.columns[df.columns.str.contains('user_id',case=False)],axis = 1,inplace = True)
df.head()
```

```python
df["post_text"] = df["post_text"].apply(lambda x: " ".join(x.lower() for x in x.split()))

# # Remove numbers from tweets
# df["post_text"] = df["post_text"].str.replace("\d","")

# # Remove punctuation from tweets
# df["post_text"] = df["post_text"].str.replace("[^\w\s]","")
df.head()
```

```python
df['post_created'] = pd.to_datetime(df['post_created'], infer_datetime_format = True)
df['weekday'] = df['post_created'].dt.weekday 
print(df[['post_created', 'weekday']].dtypes)
print(df[['post_created', 'weekday']])
```

```python
import datetime as dt
df['post_created'] = pd.to_datetime(df['post_created'], format='%a %b %d %H:%M:%S +%f %Y')

# Separate into individual columns
df['day'] = df['post_created'].dt.day
df['month'] = df['post_created'].dt.month
df['year'] = df['post_created'].dt.year
df['hour'] = df['post_created'].dt.hour
df['minute'] = df['post_created'].dt.minute
df['second'] = df['post_created'].dt.second
df.drop("post_created", axis=1, inplace=True)
df.head()
```

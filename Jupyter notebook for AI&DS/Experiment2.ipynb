```python

from matplotlib.gridspec import GridSpec
p8 = sns.kdeplot(data=pokemon, x='HP', hue='Legendary', fill=True, alpha=0.5)
p8.set(xlabel='HP', ylabel='Density')
p8.legend_.remove()

sns.despine()
p9 = sns.kdeplot(data=pokemon, x='Attack', hue='Legendary', fill=True, alpha=0.5)
p9.set(xlabel='Attack', ylabel='Density')
p9.legend_.remove()

sns.despine()

p10 = sns.kdeplot(data=pokemon, x='Defense', hue='Legendary', fill=True, alpha=0.5)
p10.set(xlabel='Defense', ylabel='Density')
p10.legend_.remove()

sns.despine()
p11 = sns.kdeplot(data=pokemon, x='Sp. Atk', hue='Legendary', fill=True, alpha=0.5)
p11.set(xlabel='Special Attack', ylabel='Density')
p11.legend_.remove()

sns.despine()

p12 = sns.kdeplot(data=pokemon, x='Sp. Def', hue='Legendary', fill=True, alpha=0.5)
p12.set(xlabel='Special Defense', ylabel='Density')
p12.legend_.remove()

sns.despine()

p13 = sns.kdeplot(data=pokemon, x='Speed', hue='Legendary', fill=True, alpha=0.5)
p13.set(xlabel='Speed', ylabel='Density')
p13.legend_.remove()

sns.despine()

# Density plot of Total
p14 = sns.kdeplot(data=pokemon, x='Total', hue='Legendary', fill=True, alpha=0.5)
p14.set(xlabel='Total', ylabel='Density')

sns.despine()

# Arrange subplots in a grid
fig = plt.figure(figsize=(12, 12))
gs = GridSpec(3, 3, height_ratios=[1, 1, 0.2], hspace=0.4)

ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[0, 2])
ax4 = fig.add_subplot(gs[1, 0])
ax5 = fig.add_subplot(gs[1, 1])
ax6 = fig.add_subplot(gs[1, 2])
ax7 = fig.add_subplot(gs[2, :])

sns.kdeplot(data=pokemon, x='HP', hue='Legendary', fill=True, alpha=0.5, ax=ax1)
sns.kdeplot(data=pokemon, x='Attack', hue='Legendary', fill=True, alpha=0.5, ax=ax2)
sns.kdeplot(data=pokemon, x='Defense', hue='Legendary', fill=True, alpha=0.5, ax=ax3)
sns.kdeplot(data=pokemon, x='Sp. Atk', hue='Legendary', fill=True, alpha=0.5, ax=ax4)
sns.kdeplot(data=pokemon, x='Sp. Def', hue='Legendary', fill=True, alpha=0.5, ax=ax5)
sns.kdeplot(data=pokemon, x='Speed', hue='Legendary', fill=True, alpha=0.5, ax=ax6)
sns.kdeplot(data=pokemon, x='Total', hue='Legendary', fill=True, alpha=0.5, ax=ax7)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python

# Assuming your DataFrame is named 'pokemon'
generation_totals = pokemon.groupby('Generation')['Total'].mean().reset_index()

# Plotting with Matplotlib
plt.figure(figsize=(8, 6))
plt.plot(generation_totals['Generation'], generation_totals['Total'], marker='o', color='green', label='Average Total')
plt.scatter(generation_totals['Generation'], generation_totals['Total'], color='red')

plt.xlabel('Generation')
plt.ylabel('Average Total')
plt.title('Average Total for each generation')
plt.legend()
plt.grid(True)
plt.show()
```

```python
generation_stats = pokemon.groupby('Generation').agg({
    'HP': 'mean',
    'Attack': 'mean',
    'Defense': 'mean',
    'Sp. Atk': 'mean',
    'Sp. Def': 'mean',
    'Speed': 'mean'
}).reset_index()
generation_stats_melted = pd.melt(generation_stats, id_vars=['Generation'], var_name='Stats', value_name='Average Stats')

plt.figure(figsize=(12, 8))
colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown']

for stat, color in zip(generation_stats_melted['Stats'].unique(), colors):
    stat_data = generation_stats_melted[generation_stats_melted['Stats'] == stat]
    plt.plot(stat_data['Generation'], stat_data['Average Stats'], marker='o', color=color, label=stat)

plt.xlabel('Generation')
plt.ylabel('Average Stats')
plt.title('Average Stats for each generation')
plt.legend(title='Stats')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
pokemon = pd.read_csv('Pokemon.csv',index_col=0)
plt.bar(pokemon['Type 1'], pokemon['Total'])
plt.xlabel('Type 1')
plt.ylabel('Total')
plt.title('Bar Graph')
plt.figure(figsize=(10,6))
plt.show()
```

```python
contingency_table = pd.crosstab(pokemon['Attack'], pokemon['Defense'])
print(contingency_table)
```

```python
import seaborn as sns
contingency_table = pd.crosstab(pokemon['Type 1'], pokemon['Type 2'])
plt.figure(figsize=(8, 6))
sns.heatmap(contingency_table, annot=True, cmap='viridis', fmt='d', cbar=True)
plt.xlabel('Type 1')
plt.ylabel('Type 2')
plt.title('Contingency Table Heatmap')
# Display the plot
plt.show()
```

```python
type_counts = pokemon['Type 1'].value_counts()
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightgreen', 'coral'])
plt.title('Type 1 concentration')
plt.figure(figsize=(8, 8))
plt.show()
```

```python
type_counts = pokemon['Type 2'].value_counts()
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightgreen', 'coral'])
plt.title('Type 2 concentration')
plt.figure(figsize=(8, 8))
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 8)) 
type1_counts = pokemon['Type 1'].value_counts()
axes[0].pie(type1_counts, labels=type1_counts.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightgreen', 'coral'])
axes[0].set_title('Type 1 Concentration')
type2_counts = pokemon['Type 2'].value_counts()
axes[1].pie(type2_counts, labels=type2_counts.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightgreen', 'coral'])
axes[1].set_title('Type 2 Concentration')
plt.tight_layout()
plt.show()
```

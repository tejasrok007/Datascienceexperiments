```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

df = pd.read_csv('package_tourism.csv')
df
```

```python
df['Places'] = df[['Place_Tourism1', 'Place_Tourism2', 'Place_Tourism3', 'Place_Tourism4', 'Place_Tourism5']].apply(lambda x: ' '.join(x.dropna()), axis=1)

# Vectorize the places using TF-IDF
vectorizer = TfidfVectorizer()
places_tfidf = vectorizer.fit_transform(df['Places'])

# Compute pairwise cosine similarity
cosine_sim = cosine_similarity(places_tfidf, places_tfidf)
```

```python
def recommend_places(place_index, cosine_sim=cosine_sim):
    sim_scores = list(enumerate(cosine_sim[place_index]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:6]  # Exclude the first place itself
    place_indices = [i[0] for i in sim_scores]
    return df.iloc[place_indices]['Places']
```

```python
place_index = 96 # Change this to the index of the place you want recommendations for
```

```python
print("Top 5 recommended places similar to", df['Places'][place_index], "are:")
print(recommend_places(place_index))
```
